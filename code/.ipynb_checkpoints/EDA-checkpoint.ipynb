# Data Examination/Cleaning

```python
# looking at null values and the categories
#some columns have so little data they can probably be ignored, Ally, PoolQC, and Misc Feature
train_df.info()
```

```python
# utilities seem like they would be important to house value
train_df['Utilities'].unique()
```

```python
# possible important value looks like categories
train_df['Neighborhood'].unique()
```

```python
# checking categories for the nan
train_df['Garage Qual'].unique()
```

```python
train_df['Bsmt Qual'].unique()
```

```python
# going to look at these as categories so just replacing the nan with a filler to be used as a category
# filling the batroom NaN values with 0 to make sure the bathroom values only has whole numbers
#The number of bathrooms should be important as a total
# looking at average room size as room size tends to be popular for home buyers
# dropping the columns with a lot of missing data
# filling the numeric columns with the mean for the missing values
# filling non numeric with 'NA' as a place holder in case I use them as categories 
def fill_in_na(data):    
    bathrooms = {'Full Bath':0, 'Half Bath':0, 'Bsmt Full Bath':0, 'Bsmt Half Bath':0}
    data.fillna(value=bathrooms, inplace=True)
    data['bath_abv'] = (data['Full Bath'] + data['Half Bath'])
    data['bath_bsmt'] = (data['Bsmt Full Bath'] + data['Bsmt Half Bath'])
    data['roomratio'] = ((data['Full Bath'] + data['Half Bath']) / data['TotRms AbvGrd'])
    data['rm_area_intr'] = data['Gr Liv Area'] * data['TotRms AbvGrd']
    data.drop(columns =['Alley','Fireplace Qu','Pool QC','Fence','Misc Feature'], inplace=True)
    data.fillna(train_df.mean(), inplace=True)
    data.fillna('NA', inplace=True)
    return data
```

Created a function for all the important changes to the training data so that I can use the same function to make sure I have those changes in the test data 

```python
fill_in_na(train_df)
```
